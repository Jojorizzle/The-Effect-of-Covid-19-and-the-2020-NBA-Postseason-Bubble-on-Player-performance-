```python
#import libraries
import re
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Scraping ESPN for Player Data
year= '2020'
page = [1,2,3,4]
dff = pd.DataFrame()
dfy = pd.DataFrame()
for i in range(len(year)):
    
    url= 'http://www.espn.com/nba/hollinger/statistics/_/page/{}/year/2020/seasontype/3'.format(page[i])

    html = urlopen(url)
    soup = BeautifulSoup(html)
    print(soup)
    headers = [th.getText() for th in soup.findAll('tr')[1].findAll('td')] 
    headers = headers[:]
    print(headers[:])
    rows = soup.findAll('tr', class_ = lambda table_rows: table_rows != "thead")
    player_stats = [[td.getText() for td in rows[i].findAll('td')]
                        for i in range(len(rows))]
    player_stats = player_stats[1:]
    df = pd.DataFrame(player_stats, columns = headers)
    df=df.drop((df.loc[df['PLAYER'] =='PLAYER']).index)
    df=df.drop(['RK', 'GP', 'MPG', 'AST', 'TO', 'USG', 'ORR', 'DRR', 'REBR', 'VA', 'EWA'],axis=1)
    dff=pd.concat([dff,df])
    dfy=year*50
```

```python
dff=dff.reset_index(drop=True)
dff.to_excel("bubbleStats.xlsx")  
```
